```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Load the cleaned data
df = pd.read_csv("Cleaned.csv")

# Handle missing or placeholder values
df.replace({"Empty": None, "Unknown": None}, inplace=True)

# Define ordinal encoding for ranked categorical features
ordinal_mappings = {
    "Cleaning Frequency": {"Never": 0, "Rarely": 1, "Monthly": 2, "Weekly": 3, "Daily": 4},
    "Bedtime": {"Early (Before 10PM)": 0, "Moderate (10PM - 12AM)": 1, "Late (After 12AM)": 2},
    "Party Frequency": {"Never": 0, "Rarely": 1, "Occasionally": 2, "Frequently": 3},
    "Personal Space": {"Minimal": 0, "Moderate": 1, "A lot": 2},
}

# Apply ordinal encoding
for col, mapping in ordinal_mappings.items():
    df[col] = df[col].map(mapping)

# One-hot encoding for categorical columns
categorical_cols = ["Gender", "Course of Study", "Year of Study", "Social Preference", "Roommate Host"]
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# One-hot encoding hobbies while preserving NaN
hobby_cols = ["Hobby 1", "Hobby 2", "Hobby 3"]
hobbies_df = pd.get_dummies(df[hobby_cols].stack()).groupby(level=0).sum()
df = df.drop(columns=hobby_cols).join(hobbies_df)

# Fill NaN values (e.g., roommate names)
df.fillna({"Roomate Full Name": "No Roommate Assigned"}, inplace=True)

# Save the encoded dataset
df.to_csv("encoded.csv", index=False)

print("Encoding complete. File saved as 'encoded.csv'.")
```
